# 串列Lists

串列的概念相當於python中的序列，它與string不同的地方在於，串列的內容是可變的，所以在串列中的每個元素都是可以調整的。

串列是由[]所建構而成，每個元素由逗號分隔。

```python
# 建立一個名為my_list的串列
my_list = [1,2,3]
```

上述範例建立了元素皆為整數的串列，但串列的元素也可以是不同類型，舉例如下：

```python
my_list = ['A string',23,100.232,'o'] #包含字串、整數、浮點數、字元
```

和字串一樣，len計算出串列中的元素個數。

```python
len(my_list)
```

### 索引與分片Indexing and Slicing
用法與字串相似，範例如下:

```python
my_list = ['one','two','three',4,5]
```

```python
# 取出索引值為0的元素
my_list[0]
```

```python
# 取出索引值1之後的所有元素
my_list[1:]
```

```python
# 取出索引值3之前的所有元素，但不包含索引值3的元素
my_list[:3]
```

也可以使用+來結合串列。

```python
my_list + ['new item']
```

請注意：上述的動作並未真的將my_list的內容改掉!

```python
my_list
```

請要覆寫mylist內容的話請用下方的語法:

```python
# Reassign
my_list = my_list + ['add new item permanently']
```

```python
my_list
```

也可以使用 * 來複製串列

```python
# 複製2次
my_list * 2
```

```python
# 請注意上述的動作並未改到原來的內容。
my_list
```

## 串列基本方法Basic List Methods

和其他程式語言相比，串列相當於陣列，但python中的串列長度不固定，所以使用起來更有彈性。

```python
# 建立一個新的串列
list1 = [1,2,3]
```

使用 **append** 方法來永久增加串列中的元素，每次加至串列的最末端:

```python
# Append
list1.append('append me!')
```

```python
# Show
list1
```

使用 **pop** 將串列中的元素取出，預設是取出最後一個索引值中的元素，但也可以指定要取出的索引值中的元素:

```python
# 將索引值0的元素取出，這是queue佇列的做法，最先進入串列的元素會先拿出來，至於會運用在什麼地方，以後會有範例說明。
list1.pop(0)
```

```python
# 看看有什麼變化
list1
```

```python
# 不指定取出元素的索引值，預設則會取出索引值-1的元素。這是stack堆疊的做法，最後進入串列的元素會最先拿出來，以後會有範例說明。
popped_item = list1.pop()
```

```python
popped_item
```

```python
# Show remaining list
list1
```

如果指定不存在元素的索引值的話，會出現錯誤，如下例所示範:

```python
list1[100]
```

使用 **sort** 和 **reverse** 來處理串列:

```python
new_list = ['a','e','x','b','c']
```

```python
#Show
new_list
```

```python
# 反轉串列元素 (會永久改變內容，使用時請注意)
new_list.reverse()
```

```python
new_list
```

```python
# 串列排序 (依照字母順序排列,但如果元素內容為數字型別則會由小到大排序)(效果同樣是永久改變內容)
new_list.sort()
```

```python
new_list
```

```python
num_list=[2353,4564,23,564,64,3465]
```

```python
num_list.sort()
```

```python
num_list
```

## 巢狀串列Nesting Lists
就是lists中還有lists

```python
# 建立三個串列
lst_1=[1,2,3]
lst_2=[4,5,6]
lst_3=[7,8,9]

# 將三個串列組合起來變成一個矩陣
matrix = [lst_1,lst_2,lst_3]
```

```python
# Show
matrix
```

現在就變成二維串列了

```python
# 拿出串列中的第一個元素
matrix[0]
```

```python
# 拿出第一個元素中的第一個元素
matrix[0][0]
```

# List Comprehensions
我不會翻譯，就是可以快速產生串列的語法

```python
# 搭配for迴圈和索引值，最後用[]組合而成的串列
first_col = [row[0] for row in matrix]
```
