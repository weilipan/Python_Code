## Statement 命題

A statement is a sentence that is either true or false.

一個可以判斷真偽的句子，我們稱為命題。

### for examples: ###

<ul>
    <li>19>2</li>
    <li>97是質數</li>
    <li>地球繞著太陽公轉</li>
    <li>月球繞著地球公轉</li>
</ul>

## and(且) 和 or(或)

<ol>
    <li>以大寫字母代表命題</li>
    <li>以符號 &#8745; 代表「且」，就是python語言中的「and」</li>
    <li>以符號 ∪ 代表「或」，就是python語言中的「or」</li>
    <li>命題與命題以 ∩  ∪ 互相連接，形成新命題</li>
</ol>

### for examples:###

A=早餐吃燕麥粥<br/>
B=早餐吃吐司<br/>
A ∩ B = 早餐吃燕麥粥和吐司<br/>
A ∪ B = 早餐吃燕麥粥或吐司<br/>

## DeMorgan's Law 笛摩根定理 ##

我們寫貪食蛇去判斷貪食蛇是否死亡，可以用以下兩個條件來判斷:<br/>
<ol>
<li>A=頭的位置在邊界</li>
<li>B=頭的位置在蛇身體上面</li>
</ol>

若貪食蛇要活著，則要滿足:
### (A'∩ B'):頭不能撞到牆，也不能撞到自己身體。 ###

根據DeMorgan's Law也可以寫成

### (A ∪ B)':不能夠(頭撞到牆或撞到自己身體) ###

### A→B=B'→A'###
若小陳是哥哥，則小陳是男生

可以推導出：若小陳不是男生，則他一定不會是哥哥。
